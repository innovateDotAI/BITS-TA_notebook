#Introduction

Graphs are data structure which can be used to represent complex system. This data structure is defined as collection of objects which is referred as nodes and their interaction as edges. Graphs are used in various areas like social network where users can be represented as nodes and relation among user as edges. In biological system we can use graph to represent protein as nodes and use the edges to represent various biological interactions, such as kinetic interactions between proteins.

**Graph Data Structure representation**:G = (V,E) where V represents nodes and E represents edges. The Order of Graph is define as number of vertices ( Referred as cardinality of vertices) and Size of graph is number of edges(E) in the graph( Referred as cardinality of Edges)

Graph Classification is method though which we try to label the unlabelled graph. In graph labelling can happen at various level like:

1.	Node level classification: For the given large graph it may happen that some node labels are missing. In this case by using various machine learning methods, we can label the graph nodes.
2.	Relation classification: In the given graph it may happen that label of edges missing. In this case we can label the edges which is called relation classification.
3.	Graph classification: It may happen in group of graphs, it may happen that some of graph label is missing. We can use various machine learning approach to label the graphs.


In this notebook graph classification method has been used.

When we would like to study graph level feature detection then we want a feature that characterize the structure of entire graph. For this we use kernel methods. Kernel methods are function which measure similarity between two data points or graph.

$ Kernel K(G,G')   $ means kernel is in real number space. For this we take inner product or dot product of $ (G.G')$ and kernel function(Matrix) should be semi-definite ( eigen value must be positive or few will be zero). As of this consequence the kernel matrix has to be symmetric matrix. Kernel has important property where there exists a feature representation $ (.) $ such that dot product of two vector will be equal to kernel matrix value.
$ K(G,G') = (G^T).(G')$ . This mean we can find such feature representation of graph which can be used as graph kernel. We do not need to define $$ explicitly as once kernel is defined, we can use kernel SVM kind of algorithms.

since kernel is similarity between two dataset, so we can define kernel in various ways like graphlet kernel, Weisfeiler -Lehman kernel, Random Walk kernel, shortest path graph kernel etc. In each kernel we see a specific kind of similarity measured between the graphs.

**Graphlet Kernel**: Like bag-of-words vector representation of given document in NLP, we can use similar concept in defining feature representation of (G). For this we define the feature vector dimension as the possible graphlets that we would like to consider in given graph. Let k=3 means graphlet size with 1 triangle, 2 two node connected and 3 no node connected. We will search this pattern in either graph and put frequency count against each shape.
```python
# Graph Generation from edge list
def graphGen(edge_list,vis):
    G = nx.Graph()
    G.add_edges_from(edge_list)
    if vis == True:
      pos = nx.spring_layout(G)  # Seed layout for reproducibility
      nx.draw(G, pos, node_color="y", node_size=300, with_labels=True)
      plt.show()
    #G.clear()
    return G
```

```python
    # Initial hashvalue generation for nodes
    def initialHashVal(Node_list):
      global hashKeystore
      global hashKeyVal
      for nod in Node_list: # Initial hash value for all nodes
        if hashKeystore.get(nod) == None:
          hashKeystore[nod] = hashKeyVal
          hashKeyVal+=1
```

```python
# Edge List generation from Graph
def edgeListGen(H):
  s = list(nx.generate_edgelist(H, data=False)) # Edges generated as strings only.
  edge_list=list()
  #print("Raw edge after mapping:",s)
  for edge in s:
    e_list = edge.split()
    # e_list=list()
    # e_list.append(edge[0])
    # e_list.append(edge[2])
    edge_list.append(e_list)
  return edge_list
# Update Edge List for next iteration , updates social_net_temp
def edgeListUpdate(hashKeystore,Node_list,g_temp,social_net_temp,gID):
  nMap = dict()
  for node in sorted(Node_list):
    for key1, val1 in hashKeystore.items():
      if str(node) in str(key1):
        nMap[node]=val1
  #print("Initial Key Mappring",nMap)
  # Update Graph by relable with hash value
  H = nx.relabel_nodes(g_temp, nMap)
  # Generate NodeList dictionary of each graph sequence 
  nList1 = list(H.nodes)
  # Update Edge List
  #print("OldEdgeList",edge_list)
  edge_list = edgeListGen(H)
  #print("NewEdgeList",edge_list)
  #print("OldSocialNEt:",social_net)
  social_net_temp[gID] = edge_list
  #print("NewSocialNEt:",social_net)
  #H.clear()
  return social_net_temp,nList1
```

```python
# Final Graph Sequence generation and KeyStoreUpdate and social_net edge list update after all iteration
def keyStoreUpdate(g1,depth,edge_list,social_net_temp):
  global hashKeystore
  global hashKeyVal
  Node_list = list(g1.nodes) # Node List of graph sequence
  nMap = dict()
  for source in sorted(Node_list): # Iteration over entier node list of graph
    #source = 16
    #print("Soruce:",source)
    allNnode = list() # All neighbourhood node list for given soruce node
    for d in range(depth+1):
      nebh_node = list(nx.descendants_at_distance(g1, source, d))  # Neighbourhodd node list
      if d ==0:
        sNode = nebh_node[0]
        soruce_node = str(nebh_node[0])+'_'
      else:
        allNnode = [*nebh_node,*allNnode] # Neighbourhodd node list # unpack both iterables in a list literal

    #print(sorted(allNnode))
    Node_key = [soruce_node,*sorted(allNnode)] # Node list for Hash value assignment

    #nMap = dict()
    key = ''.join([str(elem) for elem in Node_key])
    #print(key)
    if hashKeystore.get(key) == None:
      hashKeystore[key] = hashKeyVal
      nMap[sNode] = hashKeyVal
      hashKeyVal+=1
    else:
      nMap[sNode] = hashKeystore.get(key)
    #break
  #print('hashKeystore=',hashKeystore)
  #print('nMap=',nMap)

  # Update Graph by relable with hash value
  H = nx.relabel_nodes(g1, nMap)
  # Generate NodeList dictionary of each graph sequence 
  nList1 = list(H.nodes)
  # Update Edge List
  #print("OldEdgeList",edge_list)
  edge_list = edgeListGen(H)
  #print("NewEdgeList",edge_list)
  #print("OldSocialNEt:",social_net)
  social_net_temp[gid] = edge_list
  #print("NewSocialNEt:",social_net)
  #break
  #H.clear()
  return social_net_temp,nList1
```

```python
def featureVector(f_node,hashKeystore):
  #Feature Vector generation FM

  #Generate Graph ID from f_nodes dictionary
  gid_list = list()
  for gSeq,nList in f_node.items():
    gid_list.append(gSeq.split('_')[0])
  gList= (list(set(gid_list))) # Graph ID List 
  # Club all the nodes generated in the process of graph sequence into single list for wach graph ID
  ndict = dict()
  for each in gList:
    fList = list()
    for key in f_node:
      if key.startswith(each):
        fList.append(f_node[key])
    ndict[each] = fList # Graph ID wise node list arranged


  # Feature Vector generation from Graph sequecne using and hash values
  allNList = list(hashKeystore.values())
  fVector = list()
  fcnt = 0
  for k, v in ndict.items():
    #print("Graph ID:",k)
    #print("Node List of each graph sequence:",v)
    # Initialize feature vecotr for each graph
    fDict=dict() # Feature Vectore of individual Graphs
    fDict['GID']=k
    for n in allNList:
      fDict[n]=0
    for i in v:
      Notmatch_list = list(set(allNList)-set(i))
      diffList = list(set(i)-set(allNList)) # Check if graph has some node which is not present in allNList
      if len(diffList) ==0:
        match_list = i
      else:
        print("Logic error in Node hashing, check the program accuracy")
      for m in match_list:
        fDict[m]=fDict[m]+1
      for nm in Notmatch_list:
        fDict[nm]=fDict[nm]+0
    #print(fDict)
    fVector.append(fDict) # Feature Vector for all graphs
    fcnt+=1 # Feature vectore insertion count
    chkpnt = list(range(10,1000000,50))
    if fcnt in chkpnt:
      print("No. Featureset Inserted at",ct(),'=',fcnt)
  #print(fVector)
  return fVector
```
